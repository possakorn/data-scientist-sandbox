```python
import numpy as np
print(f'NumPy version: {np.__version__}')
```

## 1. Creating Arrays

```python
# From Python lists
a1 = np.array([1, 2, 3, 4, 5])
a2 = np.array([[1, 2, 3], [4, 5, 6]])
print('1-D:', a1)
print('2-D:\n', a2)
```

```python
# Built-in factory functions
print(np.zeros((3, 4)))
print(np.ones((2, 3)))
print(np.eye(4))
print(np.arange(0, 20, 2))        # [0, 2, 4, ... 18]
print(np.linspace(0, 1, 6))       # 6 evenly-spaced points in [0, 1]
```

## 2. Array Attributes

```python
x = np.array([[1., 2., 3.], [4., 5., 6.]])
print('shape :', x.shape)
print('ndim  :', x.ndim)
print('size  :', x.size)
print('dtype :', x.dtype)
print('nbytes:', x.nbytes)
```

## 3. Indexing & Slicing

```python
arr = np.arange(12).reshape(3, 4)
print(arr)
print('Row 0      :', arr[0])
print('Col 2      :', arr[:, 2])
print('Sub-matrix :\n', arr[0:2, 1:3])

# Boolean indexing
print('Values > 6 :', arr[arr > 6])
```

## 4. Reshaping & Stacking

```python
a = np.arange(6)
print('Original  :', a)
print('Reshaped  :\n', a.reshape(2, 3))
print('Flattened :', a.reshape(2, 3).ravel())

b = np.array([10, 20, 30])
c = np.array([40, 50, 60])
print('hstack :', np.hstack([b, c]))
print('vstack :\n', np.vstack([b, c]))
```

## 5. Vectorised Operations & Broadcasting

```python
v = np.array([1., 2., 3., 4.])
print('v * 3       :', v * 3)
print('v ** 2      :', v ** 2)
print('sqrt(v)     :', np.sqrt(v))
print('sum         :', v.sum())
print('mean        :', v.mean())
print('std         :', v.std())

# Broadcasting example: add a column vector to each column of a matrix
mat = np.ones((3, 4))
col = np.array([[1], [2], [3]])   # shape (3, 1)
print('\nBroadcast add:\n', mat + col)
```

## 6. Linear Algebra

```python
A = np.array([[2., 1.], [5., 3.]])
B = np.array([[1., 2.], [3., 4.]])

print('A @ B (matmul):\n', A @ B)
print('Determinant of A :', np.linalg.det(A))
print('Inverse of A:\n', np.linalg.inv(A))

# Eigenvalues / eigenvectors
vals, vecs = np.linalg.eig(A)
print('Eigenvalues :', vals)
print('Eigenvectors:\n', vecs)
```

## 7. Random Number Generation

```python
rng = np.random.default_rng(seed=42)

# Distributions
uniform   = rng.uniform(0, 1, size=5)
normal    = rng.normal(loc=0, scale=1, size=5)
integers  = rng.integers(1, 100, size=10)

print('Uniform :', uniform)
print('Normal  :', normal)
print('Integers:', integers)

# Shuffle & choice
arr = np.arange(10)
rng.shuffle(arr)
print('Shuffled:', arr)
print('Choice  :', rng.choice(arr, size=3, replace=False))
```
