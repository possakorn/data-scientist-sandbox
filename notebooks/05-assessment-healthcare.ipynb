```python
# import dataset
df = pd.read_csv("/Users/tytatee/Library/CloudStorage/GoogleDrive-possakorn.k@gmail.com/My Drive/00_script/c_github/data-scientist-sandbox/ds_sandbox/ds/healthcare_dataset.csv")
print(df.shape)
print(df.head(5))
df.info()
df.describe()
```

```python
# change type
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])

df['length_of_stay'] = df['Discharge Date'] - df['Date of Admission']
df
```

```python
cat_cols = ["Gender", "Blood Type", "Medical Condition", "Admission Type", "Medication", "Test Results", "Admission Type", "Insurance Provider"]

for col in cat_cols:
    counts = df[col].value_counts().reset_index()
    counts.columns = [col, "count"]
    fig = px.bar(counts, x=col, y="count", title=f"Count by {col}")
    fig.show()
```

```python
num_cols = ["Billing Amount", "length_of_stay", "Age"]

for col in num_cols:
	fig = px.histogram(df, x=col, title=f"Distribution of {col}")
	fig.show()
```
